```python
import pandas as pd
```

```python
aluguel = pd.read_excel('/content/houses_rent.xlsx')
```

```python
aluguel.head(10)
```

```python
aluguel= aluguel.rename(columns={"parking spaces":"parking_spaces","rent amount":"rent_amount","property tax":"property_tax","fire insurance":"fire_insurance" })
```

```python
aluguel.head(10)
```

```python
aluguel.dtypes
```

```python
aluguel = aluguel.astype({"rent_amount":"float","property_tax":"float","fire_insurance":"float","total":"float"})
aluguel.dtypes
```

```python
aluguel.head()
```

```python
aluguel = aluguel[(aluguel['total'])<=4000]
aluguel.head()
```

```python
aluguel.describe()
```

```python
aluguel = aluguel[(aluguel['total']<=4000) & (aluguel['animal']=='accept')]
aluguel.head()
```

```python
aluguel = aluguel[(aluguel['total']<=4000) & (aluguel['animal']=='accept')&((aluguel['rooms']==2)|(aluguel['rooms']==3))]
aluguel.head()
```

```python
aluguel = aluguel.query("total <=4000 and animal =='accept'")
aluguel.head()
```

```python
aluguel = aluguel.query(("total <=4000 and animal =='accept'and rooms == 2 or rooms ==3"))
aluguel.head()
```

```python
aluguel.describe()
```

```python
agg_city_price = aluguel.groupby(["city"])["total"].agg("mean").reset_index()
agg_city_price
```

```python
agg_bath_room_price = aluguel.groupby(["bathroom","rooms"])["total"].agg('mean').reset_index()
agg_bath_room_price
```

```python
import seaborn as sns
```

```python
ax = sns.barplot(agg_bath_room_price, x = "rooms", y="total", color = 'green');
ax.set(xlabel='Qtd quartos', ylabel='Média de preço');
```

```python
ax1 = sns.barplot(agg_bath_room_price, x = "bathroom", y="total", color = 'cyan');
ax1.set(xlabel='Qtd banheiros', ylabel='Média de preço');
```
